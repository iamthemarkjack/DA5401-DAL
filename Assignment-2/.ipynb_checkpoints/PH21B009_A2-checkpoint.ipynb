```python
import pandas as pd
import numpy as np

# Sklearn imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Plotting imports
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part - A

### Loading the data

```python
data = pd.read_csv('./mushroom_data/mushrooms.csv')
data.info()
```

```python
data.isna().sum()
```

- No Null values are there, therefore no need to fill Na values

- We will create one-hot vector for all categroical columns

```python
data.head(2)
```

```python
# getting the categorical columns
df = data.copy()
cat_cols = data.columns.tolist()
cat_cols.remove('class')

# One Hot Encoding
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_ohe = enc.fit_transform(df[cat_cols])

# creating the dataframe with one-hot encoded columns
ohe_cols = enc.get_feature_names_out(cat_cols)
df_ohe = pd.DataFrame(X_ohe, columns=ohe_cols, index=df.index)

# concatenating the one-hot encoded columns to the original dataframe
df_encoded = pd.concat([df.drop(columns=cat_cols), df_ohe], axis=1)
```

```python
# Splitting the data into features and target
X = df_encoded.drop(columns=['class'])
Y = df_encoded['class'].apply(lambda x: 1 if x == 'e' else 0)  # converting to binary
```

```python
# Dimension of the dataset before and after encoding
print("Shape of X before encoding (Including target):", df.shape)
print("Shape of X after encoding (Excluding target):", X.shape , " - With target it will be 118 dimensions")
```

```python
""" 

- One hot encoding is better than label encoding for categorical variables as it avoids ordinal relationships.
- Here, since all features are categorical, one-hot encoding is appropriate to convert them into a column of binary vectors.
- Before PCA, we will standardize the data to have mean 0 and variance 1.
- Onehot enconding is important as PCA works on numerical data.
- Also, it is better than label encoding as it avoids ordinal relationships. (as PCA is based on variance and covariance)

"""
```

### standardization

- Standarad scaler is important even on binary feat as PCA works based on Cov matrix

- columns with high variance can dominate cov matrix and can highly influence the principal components

- so, scaling the var to 1 and mean to 0 is important

- Even for binary columns, the columns with high occurances of 1 and high variance can significantly affect the PCA's principal components

- So, even for binary, it is a good practice to do standarscaler

```python
# standardization
scaler = StandardScaler()

# splitting the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# applying standardization (standardizing only using train set statistics)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Part - B (Principal component Analysis)

```python
# Fit PCA on your standardized, one-hot encoded training data
pca = PCA()
pca.fit(X_train_scaled)

# Extract explained variance metrics
evr = pca.explained_variance_ratio_
cev = np.cumsum(evr)

# Optimal number of components to explain 95% variance
optimal_n_components = np.argmax(cev >= 0.95) + 1
print(f'Optimal number of components to explain 95% variance: {optimal_n_components}')

# Scree plot: explained variance ratio (bars) + cumulative (line)
fig, ax1 = plt.subplots(figsize=(20, 12))

# Bars for explained variance ratio
n_components = 80
components = np.arange(1, n_components + 1)
bar_container = ax1.bar(components, evr[:n_components], alpha=0.7, label=f'Explained variance ratio - for first {n_components} components', color='C0')
ax1.set_xlabel('Principal Component', size=16)
ax1.set_ylabel('Explained Variance Ratio', size=16)
ax1.set_xticks(components)
ax1.set_xlim(0.5, n_components + 0.5)

# Secondary axis for cumulative curve
ax2 = ax1.twinx()
ax2.plot(components, cev[:n_components], color='C1', marker='o', label=f'Cumulative explained variance - for first {n_components} components')
ax2.set_ylabel('Cumulative Explained Variance', size=16)

# giving the 95% mark a tick if not present
ticks = list(ax2.get_yticks())
if 0.95 not in ticks:
    ticks.append(0.95)
ticks = sorted(ticks)
ax2.set_yticks(ticks)

ax2.set_ylim(0, 1.05)

# Optional reference thresholds
ax2.axhline(0.80, color='C1', linestyle='--', linewidth=1, alpha=0.7)  # 80%
ax2.axhline(0.95, color='C1', linestyle=':', linewidth=1, alpha=0.7)   # 95%

# marking the 95% line
ax2.scatter([optimal_n_components], [0.95], color='black', s=100, zorder=5)
ax2.text(optimal_n_components, 0.95, f'59 comp - 95% Explained Variance', color='black', verticalalignment='bottom', fontsize=12)

# Combined legend
lines_labels = [ax.get_legend_handles_labels() for ax in (ax1, ax2)]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
ax1.legend(lines, labels, loc='best')

plt.title(f'PCA Scree Plot (for first {n_components} components)', size=24)
plt.tight_layout()
plt.show()
```

- From the screen graph, the optimal no of components is 59 dim for 95%

- We can consider taking **`60`** dim in total

### Visualization (Projecting into 2 dim)

```python
X_train_2d = pca.transform(X_train_scaled)[:, :2]

# scatter plot of the first two principal components (colored by class)
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_train_2d[:, 0], y=X_train_2d[:, 1], hue=Y_train, palette='Set1', alpha=0.7)
plt.title('PCA Projection onto First Two Principal Components', size=20)
plt.xlabel('Principal Component 1', size=16)
plt.ylabel('Principal Component 2', size=16)
```

```python
# Taking the first 5 components and pair plotting them (colored by class)
X_train_5d = pca.transform(X_train_scaled)[:, :5]
df_5d = pd.DataFrame(X_train_5d, columns=[f'PC{i+1}' for i in range(5)])
df_5d['class'] = Y_train.values
sns.pairplot(df_5d, hue='class', palette='Set1', diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Pairplot of First 5 Principal Components', size=20, y=1.02)
plt.show()
```

- The first 2 compnents shows distinct clusters for each class

- They are not linearly seperable, but most of the forms a cluster for each class

- very little mix is there between 2 classes in 2 pca components

- Seperability is good, but not very high

- With including more classes, we can achieve high seperability

## Part-3 (Performance Evaluation with Logistic Regression)

### Baseline

```python
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, Y_train)

print(f"Logistic Regression model trained on original data ({X_train.shape[1]} dimensions) \n")

# performance on train and test sets
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# calculating accuracy
train_report = classification_report(Y_train, Y_train_pred, output_dict=False)
test_report = classification_report(Y_test, Y_test_pred, output_dict=False)

print(f"Train Accuracy (Baseline): \n {train_report}")
# for class_label, metrics in train_report.items():
#     if class_label != 'accuracy':
#         print(f"  Class {class_label}: {metrics['f1-score']:.4f}")
print(f"Test Accuracy (Baseline): \n {test_report}")
```

### PCA

```python
# Training and evaluating Logistic Regression on PCA-reduced data (59 components)
n_pca_components = 59
pca_reducer = PCA(n_components=n_pca_components)

# transforming the data
X_train_pca = pca_reducer.fit_transform(X_train_scaled)
X_test_pca = pca_reducer.transform(X_test_scaled)

# training the model
model_pca = LogisticRegression(max_iter=1000, random_state=42)
model_pca.fit(X_train_pca, Y_train)

# evaluating the model
Y_train_pca_pred = model_pca.predict(X_train_pca)
Y_test_pca_pred = model_pca.predict(X_test_pca)

# calculating accuracy
train_report_pca = classification_report(Y_train, Y_train_pca_pred, output_dict=False)
test_report_pca = classification_report(Y_test, Y_test_pca_pred, output_dict=False)

print(f"Train Accuracy (PCA): \n {train_report_pca}")
print(f"Test Accuracy (PCA): \n {test_report_pca}")
```

**Experimentation with Just 10 components**

```python
# Training and evaluating Logistic Regression on PCA-reduced data (59 components)
n_pca_components = 10
pca_reducer = PCA(n_components=n_pca_components)

# transforming the data
X_train_pca = pca_reducer.fit_transform(X_train_scaled)
X_test_pca = pca_reducer.transform(X_test_scaled)

# training the model
model_pca = LogisticRegression(max_iter=1000, random_state=42)
model_pca.fit(X_train_pca, Y_train)

# evaluating the model
Y_train_pca_pred = model_pca.predict(X_train_pca)
Y_test_pca_pred = model_pca.predict(X_test_pca)

# calculating accuracy
train_report_pca = classification_report(Y_train, Y_train_pca_pred, output_dict=False)
test_report_pca = classification_report(Y_test, Y_test_pca_pred, output_dict=False)

print(f"Train Accuracy (PCA) - 10 comp: \n {train_report_pca}")
print(f"Test Accuracy (PCA)- 10 comp: \n {test_report_pca}")
```

### Evaluation - report

- With all dimensions inlcuded we were able to get accuracy and f1 of 1.0

- Which tells us all dimensions included able to help us predict the poisionous and edible mushrooms with 100% accuracy

- Even with pca of pca dimension 59, we were able to get accuracy and f1 of 1.0

- Which tells us there is lot of redudancy in data

- We were able to get high accuracy even with half of the dimensions (effectively with 50% of dim for each datapoint)

- PCA with ~95% variance (~59 comps) preserves perfect performance, the discarded 5% appears non-essential for classification

- Even for PCA with 10 components (~ 10% var) has got 0.95 f1 score, which tells many important details lies in the top-10 components and the remaining bit of information gets covered in the remaining components upto 59 components

- The symmetric drop across classes (both go from 1.00 to ~0.95) suggests no class‑specific degradation; macro and weighted averages track closely due to balanced supports.

- Perfect baseline indicates the dataset is highly separable after one‑hot encoding; a simple linear model can achieve near‑zero Bayes error in the original feature space.

- This data is highly sepearable and lot of redundant columns, which can be reduced with PCA
